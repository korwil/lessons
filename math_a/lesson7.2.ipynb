```python
# 4-е задание
import scipy 
import scipy.linalg  

A = np.array([[1, 2, 3], [2, 16, 21], [4, 28, 73]])

P, L, U = scipy.linalg.lu(A)

print(P)
print(L)
print(U)

print(np.dot(P.transpose(), A) - np.dot(L, U))
```

```python
# 5-е задание
# x + 2y – z = 1
# 8x – 5y + 2z = 12

import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
X, Y = np.meshgrid(X, Y)

Z1 = X + 2*Y - 1
Z2 = 6 - 4*X + 2.5*Y

def Q(x, y, z):
    return (x**2 + y**2 + z**2)

ax.plot_surface(X, Y, Q(X, Y, Z1))
ax.plot_surface(X, Y, Q(X, Y, Z2))
ax.scatter(0, 0, 0,'z',50,'red')
show()
```
