1) Как относятся друг к другу множество и последовательность? (в ответе использовать слова типа: часть, целое, общее, частное, родитель, дочерний субъект и т.д.)

Предположительно, множество отновится к последовательности как родитель к дочернему субъекту

```python
# Даны 4 последовательности. Необходимо:
# исследовать их на монотонность; 
# исследовать на ограниченность;
# найти пятый по счету член.

n = 10
a = [0] * n
# ограничен снизу 1, сверху бесконечностью
for i in range(1, n):
    a[i] = 2 ** i - i

b = [0] * n
# ограничен снизу -1, сверху 0
for i in range(2, n):
    b[i] = 1 / (1 - i)

c = [0] * n
# ограничен снизу ~0 (точнее 0.4), сверху бесконечностью
for i in range(1, n):
    c[i] = (-1) ** i + (2 * i) ** .5

d = [0] * n
# ограничен снизу 2, сверху 1
for i in range(1, n):
    d[i] = (-1) ** (2 * i) + 1 / (i ** 2)

# обрезаем лишние(откуда начинаются последовательности)
my_dict = {
    'A': a[1:],
    'B': b[2:],
    'C': c[1:],
    'D': d[1:]
}
```

```python
def monotons(set):
    if all(set[i] < set[i+1] for i in range(len(set)-1)):
        return 'Монотонно возрастает'
    elif all(set[i] > set[i+1] for i in range(len(set)-1)):
        return 'Монотонно убывает'
    else:
        return 'Последовательность не возрастает и не убывает'


for k, i in my_dict.items():
    print(f'{k} = {i}')
    print(monotons(i))
    print(f'5-й по счету элемент: {k} = {i[4]}')
    print('=======================================================================================================')
```

```python
#*На языке Python предложить алгоритм вычисляющий численно предел с точностью 
from sympy import factorial
eps = 10**-7
lim = 1
i = 1
n = 1
x0 = 1
while lim > eps:
    i += 1
    n += 1
    x1 = n / pow(factorial(n), 1 / n)
    lim = abs(x0 - x1)
    x0 = x1
print(f'n_iter: {i}')
print(f'f({n}) = {x0}')
```
